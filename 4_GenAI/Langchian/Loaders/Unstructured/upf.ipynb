### **UnstructuredURLLoader**

```python
url = ['https://www.cricbuzz.com/live-cricket-scores/139461/wi-vs-ind-52nd-match-super-8-group-1-x1-v-x3-icc-mens-t20-world-cup-2026']

loader = UnstructuredURLLoader(url, mode="elements", continue_on_failure=True)

docs = loader.load()
```

```python
len(docs)
```

```python
counter = 0

while(counter < 10):
    print(docs[counter].page_content)
    counter += 1
```

### **AsyncChromiumLoader**

```python
from langchain_community.document_loaders import PlaywrightURLLoader

urls = [
    "https://www.cricbuzz.com"
]

loader = PlaywrightURLLoader(
    urls=urls,
    remove_selectors=["header", "footer"]
)

docs = loader.aload()

print(docs)
```

```python
from langchain_community.document_loaders import AsyncChromiumLoader
import asyncio

urls = [
    "https://www.cricbuzz.com"
]

user_agent_string = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 Chrome/120.0.0.0 Safari/537.36"

loader = AsyncChromiumLoader(
    urls=urls,
    headless=True,
    user_agent=user_agent_string
)

async def main():
    docs = await loader.load()
    print(docs[0].page_content[:500])
    return docs

docs = main()
```
