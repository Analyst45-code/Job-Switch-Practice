```python
data = res
```

#### **Create Document**

```python
doc = Document()
```

#### **Title Page**

```python
title = doc.add_heading()
run = title.add_run(data['title'])
title.alignment = WD_ALIGN_PARAGRAPH.CENTER
run.bold
```

**Details of Doc**

```python
doc.add_paragraph(f"Document Type: {data['document_type']}")
doc.add_paragraph(f"Author: {data['author']}")
doc.add_paragraph(f"Date: {data['date']}")
```

#### **Executive Summary**

```python
doc.add_heading("Executive Summary", 1)
doc.add_paragraph(data["executive_summary"])
```

#### **Company Overview**

```python
doc.add_heading("Company Overview", 1)

overview = data["company_overview"]

table = doc.add_table(rows=4, cols=2)
table.style = "Table Grid"

table.rows[0].cells[0].text = "Industry"
table.rows[0].cells[1].text = overview["industry"]

table.rows[1].cells[0].text = "Employees"
table.rows[1].cells[1].text = str(overview["employees"])

table.rows[2].cells[0].text = "Annual Revenue"
table.rows[2].cells[1].text = overview["annual_revenue"]

table.rows[3].cells[0].text = "Regions"
table.rows[3].cells[1].text = ", ".join(overview["regions"])
```
