```python
from docx import Document

def save_to_docx(text, filename="output.docx"):
    document = Document()

    lines = text.split("\n")

    for line in lines:
        if line.strip().startswith("#"):
            document.add_heading(line.replace("#", "").strip(), level=1)
        elif line.strip().startswith("-"):
            document.add_paragraph(line.strip(), style="List Bullet")
        else:
            document.add_paragraph(line)

    document.save(filename)

save_to_docx(response, "rag_report.docx")
```

7️⃣ Template-Based DOCX (Best Practice)

Instead of building layout in code:

Create a .docx template in Word

Add placeholders:

{{title}}
{{summary}}
{{conclusion}}

Then replace them:

from docx import Document

def replace_placeholder(doc, key, value):
    for p in doc.paragraphs:
        if key in p.text:
            p.text = p.text.replace(key, value)

doc = Document("template.docx")

replace_placeholder(doc, "{{title}}", "AI Strategy Report")
replace_placeholder(doc, "{{summary}}", "This is the summary.")
replace_placeholder(doc, "{{conclusion}}", "Final thoughts.")

doc.save("filled_template.docx")

This is much cleaner for enterprise use.
