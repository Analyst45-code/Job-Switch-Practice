🏗 STEP 3 — Enterprise Template Engine
📄 services/template_engine.py
from docx import Document

class TemplateEngine:

    def __init__(self, template_path):
        self.doc = Document(template_path)

    def replace_text(self, placeholder, value):
        for p in self.doc.paragraphs:
            if placeholder in p.text:
                p.text = p.text.replace(placeholder, str(value))

    def replace_list(self, placeholder, items):
        for p in self.doc.paragraphs:
            if placeholder in p.text:
                p.text = ""
                for item in items:
                    self.doc.add_paragraph(item, style="List Bullet")

    def replace_table(self, placeholder, table_data, headers):
        for p in self.doc.paragraphs:
            if placeholder in p.text:
                p.text = ""
                table = self.doc.add_table(rows=1, cols=len(headers))
                table.style = "Table Grid"

                for i, header in enumerate(headers):
                    table.rows[0].cells[i].text = header

                for row_data in table_data:
                    row = table.add_row().cells
                    for i, key in enumerate(row_data):
                        row[i].text = str(row_data[key])

    def save(self, output_path):
        self.doc.save(output_path)
🧠 STEP 4 — DOC Generator (Business Logic Layer)
📄 services/doc_generator.py
from services.template_engine import TemplateEngine


def generate_ai_strategy_report(res):

    engine = TemplateEngine("templates/ai_strategy_template.docx")

    # Simple text replacements
    engine.replace_text("{{title}}", res["title"])
    engine.replace_text("{{document_type}}", res["document_type"])
    engine.replace_text("{{author}}", res["author"])
    engine.replace_text("{{date}}", res["date"])
    engine.replace_text("{{executive_summary}}", res["executive_summary"])
    engine.replace_text("{{conclusion}}", res["conclusion"])

    # Company overview table
    overview = res["company_overview"]
    overview_table = [
        {"Field": "Industry", "Value": overview["industry"]},
        {"Field": "Employees", "Value": overview["employees"]},
        {"Field": "Revenue", "Value": overview["annual_revenue"]},
        {"Field": "Regions", "Value": ", ".join(overview["regions"])}
    ]

    engine.replace_table(
        "{{company_overview_table}}",
        overview_table,
        headers=["Field", "Value"]
    )

    # Lists
    engine.replace_list("{{objectives_list}}", res["objectives"])
    engine.replace_list("{{key_findings_list}}", res["key_findings"])

    # Department table
    engine.replace_table(
        "{{department_table}}",
        res["department_analysis"],
        headers=["department", "current_state", "ai_opportunity", "estimated_roi"]
    )

    # Risk table
    engine.replace_table(
        "{{risk_table}}",
        res["risk_assessment"],
        headers=["risk", "impact", "mitigation"]
    )

    # Financial table
    finance_table = [{
        "Investment": res["financial_projection"]["estimated_investment"],
        "Savings": res["financial_projection"]["expected_savings"],
        "ROI (%)": res["financial_projection"]["roi_percentage"]
    }]

    engine.replace_table(
        "{{financial_table}}",
        finance_table,
        headers=["Investment", "Savings", "ROI (%)"]
    )

    output_file = "AI_Strategy_Report_Output.docx"
    engine.save(output_file)

    return output_file
