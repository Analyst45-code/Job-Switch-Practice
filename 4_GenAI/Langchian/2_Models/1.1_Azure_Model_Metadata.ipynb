### LangChain + Azure OpenAI Detailed Response Schema

Below is a **detailed response schema** for `AzureChatOpenAI` (chat models) when used via LangChain.  
This represents the structured object returned after invoking the model.

---

#### 1截뤴깵 Top-Level Response Object

| Field | Type | Description |
|-------|------|------------|
| `content` | string | Final generated text output |
| `additional_kwargs` | dict | Extra provider-specific data (e.g., tool calls) |
| `response_metadata` | dict | Raw model metadata (token usage, finish reason, etc.) |
| `tool_calls` | list | Tool/function calls generated by the model |
| `usage_metadata` | dict | Token usage summary |
| `id` | string | Unique run or response ID |
| `name` | string | Optional message name |
| `type` | string | Message type (`ai`) |

---

#### 2截뤴깵 `response_metadata` Schema

Contains provider-specific raw information from Azure OpenAI.

| Field | Type | Description | Example |
|-------|------|------------|----------|
| `token_usage` | dict | Detailed token breakdown | `{ "prompt_tokens": 25, "completion_tokens": 60 }` |
| `model_name` | string | Model used for generation | `"gpt-4"` |
| `system_fingerprint` | string | Model system version identifier | `"fp_abc123"` |
| `finish_reason` | string | Why generation stopped | `"stop"`, `"length"`, `"content_filter"` |
| `logprobs` | dict/null | Token log probabilities (if enabled) | `{...}` |

### `token_usage` breakdown

| Field | Type | Description |
|-------|------|------------|
| `prompt_tokens` | int | Tokens in input |
| `completion_tokens` | int | Tokens generated |
| `total_tokens` | int | Total tokens used |

---

#### 3截뤴깵 `usage_metadata` Schema (LangChain Standardized)

LangChain provides normalized usage tracking:

| Field | Type | Description |
|-------|------|------------|
| `input_tokens` | int | Input tokens |
| `output_tokens` | int | Output tokens |
| `total_tokens` | int | Total tokens |

---

#### 4截뤴깵 `tool_calls` Schema (If Function Calling Enabled)

If the model calls a tool/function:

| Field | Type | Description |
|-------|------|------------|
| `id` | string | Unique tool call ID |
| `name` | string | Tool/function name |
| `args` | dict | Parsed arguments |

Example:

```json
[
  {
    "id": "call_abc123",
    "name": "get_weather",
    "args": {
      "city": "Seattle"
    }
  }
]
