### **Azure OpenAI**

```python
from langchain_openai import AzureChatOpenAI

llm = AzureChatOpenAI(
    azure_endpoint="https://your-resource.openai.azure.com/",
    api_key="your-api-key",
    api_version="2024-02-01",
    azure_deployment="gpt-4-deployment",

    temperature=0.2,
    max_tokens=500,
    top_p=0.95,
    frequency_penalty=0,
    presence_penalty=0,
    streaming=True
)
```

#### 1️⃣ **Azure OpenAI Connection Parameters** (Required Setup)

| Parameter           | Type   | Required | Description |
|--------------------|--------|----------|-------------|
| azure_endpoint     | string | ✅       | Your Azure OpenAI endpoint URL (e.g., https://your-resource.openai.azure.com) |
| api_key            | string | ✅       | Azure OpenAI API key |
| api_version        | string | ✅       | API version (e.g., 2024-02-01) |
| azure_deployment   | string | ✅       | Deployment name you created in Azure (NOT the model name) |
| organization       | string | ❌       | Organization ID (rarely used with Azure) |
| timeout            | float  | ❌       | Request timeout in seconds |
| max_retries        | int    | ❌       | Number of automatic retries on failure |
| default_headers    | dict   | ❌       | Custom headers if needed |
| default_query      | dict   | ❌       | Custom query parameters |

---

#### 2️⃣ **Chat / Completion Model Parameters**  
*(Used in AzureChatOpenAI or AzureOpenAI in LangChain)*

| Parameter         | Type        | Default | Description |
|------------------|------------|---------|-------------|
| temperature      | float      | 0.7     | Controls randomness (0 = deterministic, 1 = creative) |
| max_tokens       | int        | None    | Maximum tokens in response |
| top_p            | float      | 1.0     | Nucleus sampling (alternative to temperature) |
| frequency_penalty| float      | 0.0     | Penalizes repeated tokens |
| presence_penalty | float      | 0.0     | Penalizes repeating topics |
| stop             | list[str]  | None    | Stop sequences |
| seed             | int        | None    | For reproducible outputs (if supported) |
| logprobs         | bool       | False   | Return log probabilities |
| streaming        | bool       | False   | Enable token streaming |
| model_kwargs     | dict       | None    | Pass additional raw OpenAI parameters |

---

#### 3️⃣ **Embedding Model Parameters**  
*(Used in AzureOpenAIEmbeddings)*

| Parameter         | Type   | Required | Description |
|------------------|--------|----------|-------------|
| azure_deployment | string | ✅       | Embedding deployment name |
| model            | string | ❌       | Model name (optional for Azure, usually inferred) |
| chunk_size       | int    | 2048     | Batch size for embedding calls |
| request_timeout  | float  | ❌       | Timeout for embedding request |
| max_retries      | int    | ❌       | Retry attempts |

---

#### 4️⃣ **Common LangChain Runtime Parameters**  
*These are not Azure-specific but commonly used in production pipelines.*

| Parameter | Type      | Description |
|-----------|----------|-------------|
| callbacks | list     | Custom callback handlers |
| verbose   | bool     | Enables logging |
| tags      | list[str]| Tags for tracing (LangSmith) |
| metadata  | dict     | Custom metadata for tracing |
| run_name  | string   | Custom run identifier |

---
#### **Azure Specefic**
| Concept                   | Important Detail                                  |
|---------------------------|--------------------------------------------------|
| Model name vs Deployment  | Azure uses deployment name, not raw model name   |
| API version               | Must match Azure-supported version               |
| Rate limits               | Controlled at Azure resource level               |
| Authentication            | API key or Azure AD token                        |
