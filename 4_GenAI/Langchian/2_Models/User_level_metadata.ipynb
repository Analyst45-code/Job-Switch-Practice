#### **LangChain `metadata` Detailed Schema**

| Field | Type | Required | Description | Example |
|-------|------|----------|------------|----------|
| metadata | dict[str, Any] | ❌ | Arbitrary key-value dictionary attached to a run for tracing, logging, filtering, and observability (e.g., in LangSmith). | `{"user_id": "123", "env": "prod"}` |

---

##### **Recommended Metadata Structure** (Best Practice Schema)

Although `metadata` accepts **any dictionary**, in production systems it's recommended to standardize fields like below:

| Key | Type | Purpose | Example |
|-----|------|----------|----------|
| `user_id` | string | Identify end user | `"u_48291"` |
| `session_id` | string | Track conversation session | `"sess_abc123"` |
| `conversation_id` | string | Logical conversation grouping | `"conv_2024_09_01"` |
| `request_id` | string | Unique request trace ID | `"req_789xyz"` |
| `environment` | string | Environment label | `"dev"`, `"staging"`, `"prod"` |
| `application` | string | Application name | `"customer-support-bot"` |
| `version` | string | App or prompt version | `"v1.4.2"` |
| `feature` | string | Feature flag or flow name | `"rag_pipeline"` |
| `tenant_id` | string | Multi-tenant identifier | `"tenant_acme"` |
| `experiment` | string | A/B test label | `"prompt_v2_test"` |
| `source` | string | Request origin | `"web"`, `"api"`, `"mobile"` |
| `region` | string | Deployment region | `"eastus"` |
