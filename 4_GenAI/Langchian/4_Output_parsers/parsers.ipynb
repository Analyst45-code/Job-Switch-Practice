### **🔹 1️⃣ Basic Output Parsers**

#### **1. StrOutputParser**

✅ Returns raw string output from the LLM.

**Use when:** You just need text.


````python
from langchain_core.output_parsers import StrOutputParser

parser = StrOutputParser()
chain = prompt | llm | parser
````

### **🔹 2️⃣ Structured Output Parsers**

#### **2. StructuredOutputParser**
✅ Enforces a structured response format using response schemas.


````python
from langchain.output_parsers import StructuredOutputParser, ResponseSchema

response_schemas = [
    ResponseSchema(name="name", description="Name of the person"),
    ResponseSchema(name="age", description="Age of the person"),
]

parser = StructuredOutputParser.from_response_schemas(response_schemas)

format_instructions = parser.get_format_instructions()

# Use in prompt:

prompt = PromptTemplate(
    template="Extract info:\n{text}\n{format_instructions}",
    input_variables=["text"],
    partial_variables={"format_instructions": format_instructions}
)
````

### **🔹 4️⃣ Pydantic Parsers (VERY IMPORTANT)**

#### **5. PydanticOutputParser**

✅ Parses output directly into a Pydantic model.


````python
from langchain_core.output_parsers import PydanticOutputParser

from pydantic import BaseModel

class Person(BaseModel):
    name: str
    age: int

parser = PydanticOutputParser(pydantic_object=Person)

# Returns: Person(name="John", age=30)
````

### 🔥 **Best for production structured output.**

#### **🔹 5️⃣ Enum / Typed Parsers**
#### **6. EnumOutputParser**

For outputs that must match predefined values.

Ensures output is one of enum values.

````python
from enum import Enum
from langchain.output_parsers import EnumOutputParser

class Sentiment(Enum):
    POSITIVE = "positive"
    NEGATIVE = "negative"

parser = EnumOutputParser(enum=Sentiment)
````

### 🔹 7️⃣ **Retry / Fixing Parsers (Production Critical)**

#### **10. OutputFixingParser**

Automatically fixes invalid output using LLM.


If JSON breaks → it retries and fixes.


````python
from langchain.output_parsers import OutputFixingParser

fixing_parser = OutputFixingParser.from_llm(
    parser=parser,
    llm=llm
)
````

---

#### 🧠 **When to Use What?**

| Use Case | Best Parser |
|-----------|-------------|
| Just text | `StrOutputParser` |
| JSON required | `JsonOutputParser` |
| Strong schema validation | `PydanticOutputParser` |
| Enum constraint | `EnumOutputParser` |
| List extraction | `CommaSeparatedListOutputParser` |
| Production reliability | `OutputFixingParser` |
| Modern tool calling | `with_structured_output()` |

---

#### 🚀 **Most Important in 2026**

In modern LangChain:

- ✅ `StrOutputParser`
- ✅ `JsonOutputParser`
- ✅ `PydanticOutputParser`
- ✅ `OutputFixingParser`
- ✅ `with_structured_output()` (preferred for OpenAI models)

Older ones like `StructuredOutputParser` are used less now.
