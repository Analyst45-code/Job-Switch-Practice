Adding column

```python
import pandas as pd
```

```python
df = pd.read_csv('../Datasets/Flavors.csv')
df.columns
```

### Column Level Arithmetic Operations

```python
# Same for others as well
new = df['Total Rating'] + df['Texture Rating']
print(new.dtype)
print(type(new))
print(new)
```

```python
# Multilevel Operations
multi = (df['Total Rating'] + df['Texture Rating']) * (1 * 11)
print(multi)
```

### 🔁 Column Creation & Insertion

```python
# Add to last 
df['Avg_Rating'] = df['Total Rating'] + 1
print(df)
```

##### Insert At specific Location

```python
# using insert function at specific place
sum = df['Total Rating'] * 0.02

if not 'SSS2' in df.columns:
    df.insert(0,'SSS2',sum)
```

```python
print(df)
```

### Renamin Column

```python
df = df.rename(columns={'Average R':'A R'})
print(df)
```

```python
# Multiple Columns rename
df = df.rename(columns={'Flavor Rating' : 'F R', 'Texture Rating':'Texture R'})
print(df)
```

#### Droping Column

```python
# df = df.drop(columns='SSS2')
if 'Total Rating' in df.columns:
    df.drop(columns='Total Rating', inplace=True)
```

```python
# Droping Multiple columns
if ('F R' in df.columns) & ('A R' in df.columns):
    df.drop(columns=['F R', 'A R'], inplace=True)

print(df)
```
