### Conditional Column

1. ✅ Using apply() with lambda

```python
conditional_col = data['Total Rating'].apply(lambda x: 'Great' if x > 8 else 'Not Great')
print(type(conditional_col))
print(conditional_col)
```

2. 🧱 Using np.where()

```python
import numpy as np
data['status'] = np.where(data['Total Rating'] >= 20, 'pass', 'fail')
```

3. 📌 Using np.select() for multiple conditions

```python

conditions = [data['Total Rating'] <= 20, data['Total Rating'] >= 20]
choices = ['A', 'B']
data['grade'] = np.select(conditions, choices, default='F')
```

4. 🧠 Using pd.cut() for binning

```python
data['Conditional Col'] = pd.cut(data['Total Rating'], bins=[0,6,8,100], labels=['F','B','A'])  # Bin scores into grades
print(data)
```

5. ⚙️ Using .loc[] for conditional assignment

* Explanation: More readable for multiple conditions or when assigning in steps.

```python
data.loc[data['Total Rating'] >= 50, 'status'] = 'pass'
data.loc[data['Total Rating'] < 50, 'status'] = 'fail'
```

7. 🧪 Using .map() with a dict (for categorical conditions)

* Explanation: Replace categorical codes with labels.

```python
data['status 2'] = data['status'].map({'pass' : 'P', 'fail': 'F'})

print(data['status 2'])
```

✅ 1. Using np.select() for Range Mapping

* ✔️ Best for: Multiple ranges, clear and fast.

```python
import numpy as np

conditions = [
    (data['Total Rating'] < 50),
    (data['Total Rating'] >= 50) & (data['Total Rating'] < 80),
    (data['Total Rating'] >= 80)
]
choices = ['Low', 'Medium', 'High']

data['score_level'] = np.select(conditions, choices, default='Unknown')

print(data)
```
