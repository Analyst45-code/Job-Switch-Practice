```python
import pandas as pd
import numpy as np

# Sample messy DataFrame
data = {
    'Name': [' Alice ', 'Bob', 'Charlie', 'Alice', None, 'Eve', 'BOB', 'George'],
    'Age': [25, -1, 35, np.nan, 40, 29, 25, 'unknown'],
    'Email': ['alice@example.com', 'bob@example', None, 'alice@example.com', 'eve@', 'eve@example.com', 'bob@example', 'george@example.com'],
    'Signup Date': ['2021-01-01', '2021-02-30', 'not available', '2021-01-01', None, '2021-06-01', '2021-06-01', '2021/07/20'],
    'Score': ['90', 'eighty', '85', 'NaN', 100, None, '90', '88'],
    'Active': ['yes', 'no', 'YES', 'n/a', 'yes', None, '', 'no']
}

df = pd.DataFrame(data)

# Display the raw DataFrame
print(df)
```

```python
# 1. Strip whitespace and standardize case in string columns like 'Name'
df['Name'] = df['Name'].str.strip()

df['Name'].head(5)
```

```python
# 2. Handle missing values in 'Name', 'Email', 'Active' columns

replace = {
    'Name' : 'Unknown',
    'Email' : 'Unknown',
    'Active':df['Active'].fillna(df['Active'].mode()[0])
}

for col, value in replace.items():
    if df[col].isna().any():
        df[col] = df[col].fillna(value)


mo = df['Active'].mode()[0]
print(mo)
df['Active'].replace('n/a', mo,inplace=True)
print(df.head(5))
```

```python
# 3. Convert 'Age' column to numeric, replacing invalid or unknown values
print(df['Age'].dtype)

df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

age_mean = round(df['Age'].mean())
print(f'Age_mean : {age_mean}')

df['Age'] = df['Age'].fillna(age_mean)
print(df['Age'].head(7))

# 4. Remove or fix negative or unrealistic 'Age' values
df.loc[(df['Age'] < 1) | (df['Age'].isna()), 'Age'] = age_mean

df['Age'] = df['Age'].astype('Int64')  # Nullable integer

print(df['Age'].head(7))
```

```python
# 5. Standardize and validate email addresses using regex

pattern = r'\w*@\D{3,9}\.com$'

print(df.loc[~df['Email'].str.contains(pattern, na=False)])

# Replace invalid emails
df.loc[~df['Email'].str.contains(pattern, na=False), 'Email'] = 'Unknown'

print(df)
```
