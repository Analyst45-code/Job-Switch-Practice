```python
import pandas as pd
import numpy as np

# Sample data for DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, np.nan, 40],
    'Salary': [50000, 60000, 55000, 52000, np.nan],
    'Department': ['HR', 'Engineering', 'HR', 'Engineering', 'Marketing']
}

df = pd.DataFrame(data)

# 1. Overview of the DataFrame structure
print("\n=== df.shape ===")
print(df.shape)                     # (5, 4) → 5 rows, 4 columns

print("\n=== df.size ===")
print(df.size)                      # 20 → 5 * 4 = 20 elements total

print("\n=== df.ndim ===")
print(df.ndim)                      # 2 → DataFrame is 2-dimensional

print("\n=== df.columns ===")
print(df.columns.tolist())          # ['Name', 'Age', 'Salary', 'Department']

print("\n=== df.index ===")
print(df.index.tolist())            # [0, 1, 2, 3, 4]

print("\n=== df.dtypes ===")
print(df.dtypes)                    # Data types of each column


print("\n=== df.memory_usage() ===")
print(df.memory_usage())            # Memory usage per column

print("\n=== df.isnull().sum() ===")
print(df.isnull().sum())            # Missing values per column
# Age: 1, Salary: 1

print("\n=== df.nunique() ===")
print(df.nunique())                 # Unique values per column
```

```python

# Numeric Functions
print("\n=== df['Department'].value_counts() ===")
print(df['Department'].value_counts())              # Frequency of each unique department

print("\n=== df.corr(numeric_only=True) ===")
print(df.corr(numeric_only=True))                   # Correlation between numeric columns

print("\n=== df.mean(numeric_only=True) ===")
print(df.mean(numeric_only=True))                    # Mean of numeric columns

print("\n=== df.median(numeric_only=True) ===")
print(df.median(numeric_only=True))  # Median of numeric columns

print("\n=== df.mode() ===")
print(df.mode())  # Most frequent value(s) per column

print("\n=== df.min(numeric_only=True) ===")
print(df.min(numeric_only=True))  # Minimum values

print("\n=== df.max(numeric_only=True) ===")
print(df.max(numeric_only=True))  # Maximum values

print("\n=== df.std(numeric_only=True) ===")
print(df.std(numeric_only=True))  # Standard deviation

print("\n=== df.var(numeric_only=True) ===")
print(df.var(numeric_only=True))  # Variance

print("\n=== df.count() ===")
print(df.count())  # Non-null count per column

print("\n=== df.sample(2) ===")
print(df.sample(2))  # Random sample of 2 rows
```
