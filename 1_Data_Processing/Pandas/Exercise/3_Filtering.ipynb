```python
import pandas as pd

data = {
    'Country': ['USA', 'China', 'Japan', 'Germany', 'India', 'Australia', 'UK'],
    'Gold Medal': [39, 38, 27, 10, 1, 17, 22],
    'Silver Medal': [41, 32, 14, 11, 2, 7, 21],
    'Bronze Medal': [33, 18, 17, 16, 1, 22, 22],
    'Total': [113, 88, 58, 37, 4, 46, 65],
    'Continent': ['America', 'Asia', 'Asia', 'Europe', 'Asia', 'Oceania', 'Europe'],
    'Rank Overall': [1, 2, 3, 4, 5, 6, 7]
}

df = pd.DataFrame(data)
```

```python
df[:2][:2]
```

1. Filter rows where Gold Medal > 20

```python
df[df['Gold Medal'] > 20][["Country", "Total", 'Gold Medal']]
```

2. Filter rows where Continent is Asia

```python
df[df['Continent'] == 'Asia']
```

3. Filter with multiple conditions

```python
df[(df['Gold Medal']> 20) & (df['Country'] == 'USA')]
```

4. Filter using isin()

```python
df[df['Country'].isin(['USA', 'India','Japan'])]
```

5. Filter using str.contains() (for column with strings)

```python
df = df[df['Country'].str.contains('i', case=False)]
df.head()
```

```python
# col = [c for c in df.columns if c.str.contains('Medal', case=False)]
# col

col2 = list(filter(lambda c: 'Medal' in c, df.columns))
col2
```

Filter countries not in Asia.

```python
df[df['Country'] != 'Asia']
```

Countries whose name starts with 'J'

```python
df[df['Country'].str.startswith('J')]
```

Keep only columns with 'Medal' in the name.

```python
medal_col = list(filter(lambda c: 'medal' in c.lower(), df.columns))
medal_col

df[medal_col]
```

Drop all columns that contain 'Rank'.

```python
rnk = list(filter(lambda c: 'rank' in c.lower(), df.columns))

df.drop(rnk, axis= 1)
```

#### **6️⃣ Using between()**

Countries with Total medals between 50 and 100.

Gold Medals between 10 and 40.

```python
# df[(df['Total'] > 50) & (df['Total'] < 100)]

df[df['Total'].between(30, 100)]
```

#### **7️⃣ Filtering with query()**

Using string query to filter:

```python
df[:].query('`Gold Medal` > 10 and `Total` > 60' )
```

```python
df.iloc[:, :6].query('`Total` > 50')
```

#### 9️⃣ **Using apply() for custom filtering**

Filter countries with Total > Gold + Silver.

Filter countries where length of Country name > 5.

```python
df[df['Country'].str.len() > 1]
```

```python
filtered_df = df[df['Total'] > (df['Gold Medal'] + df['Silver Medal'])]
filtered_df
```

#### **🔟 Bonus Challenge (Hard)**

Filter countries where Gold > Silver AND Silver > Bronze.

Filter countries whose Total is the max within their Continent (group-wise filter).

```python
df[(df['Gold Medal'] > df['Silver Medal']) & (df['Silver Medal'] > df['Bronze Medal'])]
```

```python
df[df['Total'] == df.groupby('Continent')['Total'].transform('max')]
```

```python
df[
    (df['Gold Medal'] > df['Silver Medal']) &
    (df['Silver Medal'] > df['Bronze Medal']) &
    (df['Total'] == df.groupby('Continent')['Total'].transform('max'))
]
```
