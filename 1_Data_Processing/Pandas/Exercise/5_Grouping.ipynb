#### **Notes**
- ***as_index=False*** keeps Country as a column instead of index.

```python
import pandas as pd
import numpy as np

data = {
    'Country': ['USA', 'USA', 'China', 'China', 'India', 'India', 'Germany', 'Australia'],
    'Sport': ['Athletics', 'Swimming', 'Athletics', 'Table Tennis',
              'Athletics', 'Hockey', 'Athletics', 'Swimming'],
    'Gold': [10, 12, 8, 5, 6, 4, 3, 7],
    'Silver': [8, 9, 6, 4, 5, 3, 2, 6],
    'Bronze': [7, 6, 5, 3, 4, 2, 1, 5]
}

df = pd.DataFrame(data)
df
```

#### 🧠**Grouping & Aggregation**

```python
# 1️⃣ Basic Grouping

# Group the data by Country
grouped = df.groupby(by='Country')
grouped.sum()
```

```python
# Calculate the total Gold medals for each country
groupC = df.groupby(by='Country')['Gold'].sum()
groupC


# Option 1: Select column before summing (recommended)
groupC = df.groupby('Country')['Gold'].sum()
groupC

# Option 2: Select column after grouping
groupC = df.groupby('Country').sum()[['Gold']]
groupC

# Option 3: Keep result as DataFrame
groupC = df.groupby('Country', as_index=False)['Gold'].sum()
groupC
```

#### 2️⃣ **Multiple Aggregations**

```python
# Group by Country
# Find:   Total Gold      Total Silver    Total Bronze

groupM = df.groupby(by='Country', as_index=False)
groupM.agg('sum')

# Approach 2
df.groupby('Country', as_index=False).sum()
```

#### 3️⃣ **Group by Multiple Columns**
- Group by Country and Sport
- Find the sum of all medals

```python
groupM2 = df.groupby(by=['Country', 'Sport']).sum()
groupM2
```

#### **4️⃣ Different Aggregations per Column**
- Group by Country
- Calculate:
    - Sum of Gold
    - Mean of Silver
    - Max of Bronze

```python
import numpy as np
groupMC = df.groupby(by='Country', as_index=False).agg({
    'Gold':'sum',
    'Silver': 'mean',
    'Bronze':'max'
})

groupMC
```

#### 5️⃣ **Using .agg() with Custom Names**
- Group by Country
- Create columns named:
    - Total_Gold
    - Avg_Silver
    - Max_Bronze

```python
# With Alia name
groupMC = df.groupby('Country').agg(
    Total_Gold=('Gold', 'sum'),        # alias: Total_Gold
    Average_Silver=('Silver', 'mean'), # alias: Average_Silver
    Max_Bronze=('Bronze', 'max')       # alias: Max_Bronze
).reset_index()

print(groupMC)
```

#### 6️⃣ **Sorting Aggregated Results**
- Group by Country
- Find total medals (Gold + Silver + Bronze)
- Sort countries by highest total medals

```python
df['Total'] = df['Gold'] + df['Silver'] + df['Bronze']

# group_sorted = df.groupby('Country').agg(sum).sort_values('Total', ascending=False)
# group_sorted.head(5)
# # df.head(5)

grouped = df.groupby('Country').sum().reset_index()

group_sorted = grouped.sort_values('Total', ascending=False)

print(group_sorted.head(5))
type(group_sorted)
```

#### 7️⃣ **Filtering Groups**

Show only countries with total Gold medals greater than 10


```python
group_sorted = group_sorted[group_sorted['Total']> 10]
group_sorted
```

**8️⃣ Bonus (Advanced)**
- **Group by Sport:** Find which sport has the highest average Gold medals


```python
group_sport = df.groupby('Sport')['Gold'].agg(np.average).reset_index().sort_values('Gold', ascending=False)[:1]
group_sport.head(5)
```

### 💡 **Tips**
- Use .groupby()
- Use .sum(), .mean(), .max()
- Use .agg() for flexibility
- Chain .sort_values() and .reset_index() if needed
