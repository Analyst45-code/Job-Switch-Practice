#### **Part 4: Column-based sorting tricks**


Reorder the dataframe columns so that:
- Country comes first
- Gold, Silver, Bronze, Total come next


```python
order = ['Continent', 'Gold Medal', 'Silver Medal', 'Total']

# Reorder columns
df = df[order + [c for c in df.columns if c not in order]]

df.head(4)
```

```python
cols_ordered = df.columns.get_indexer(order)
df = df.iloc[:, cols_ordered]

df.head(4)
```

Sort columns alphabetically using axis=1.

```python
df = df.sort_index(axis=1)
df.head(5)
```

Move all medal-related columns to the right side of the dataframe.

```python
medal_cols = [c for c in df.columns if any(m in c for m in ['Gold', 'Silver', 'Bronze'])]
other_cols = [c for c in df.columns if c not in medal_cols]

df = df[other_cols + medal_cols]
df.head(5)
```

Sort columns by length of column name.

```python
sorted_cols = sorted(df.columns, key=len, reverse=True)
sorted_cols

df = df[sorted_cols]
df.head()

df = df[sorted(df.columns, key=len, reverse=True)]
df.tail()
```

Sort columns so that columns containing "Rank" appear first.


```python
rnk = [c for c in df.columns if 'Rank' in c]
print(rnk)
other = [c for c in df.columns if c not in rnk]
print(other)
df = df[rnk + other]
df.head()
```
