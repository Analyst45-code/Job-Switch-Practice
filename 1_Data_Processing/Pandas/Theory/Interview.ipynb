> 4. What are the differences between `apply()`, `map()`, `applymap()`, and `transform()`?
* ✅ map() --> Used with: Series only. 
    * 🧠 Purpose: Element-wise transformation.  
    * 🛠️ Function types: Accepts a function, dictionary, or Series.

* ✅ apply() --> Used with: Both Series and DataFrame.  
    * 🧠 Purpose (Series): Applies a function to each element. (DataFrame): to each row or column. 
    * 🛠️ Axis param: axis=0 (columns), axis=1 (rows).


* ✅ applymap() --> Used with: DataFrame only.  
    *  🧠 Purpose: Element-wise transformation of entire DataFrame. 
    * 🛠️ Function types: Accepts element-wise functions only. 
    * 📌 Example: df.applymap(lambda x: x**2)  # Squares every cell

* transform()--> Used with: Series and DataFrame.   
    * 🧠 Purpose: Element-wise transformations while preserving shape. 
    * ⚠️ Important: Output must have same shape as input.
    * ✅ Common Use Case: Group-wise transformations (groupby().transform()).
    * 📌 Example: df['A'].transform(lambda x: x - x.mean())  # Centering

## ✅ Data Manipulation

### 6. How do you merge, join, and concatenate DataFrames in Pandas?

- `merge()` is used to combine DataFrames using common columns (like SQL joins).
- `join()` is a method of DataFrame to combine columns of another DataFrame based on index or key.
- `concat()` is used to stack multiple DataFrames either vertically (row-wise) or horizontally (column-wise).

---

### 7. Explain the differences between `merge()`, `join()`, and `concat()`.

- **`merge()`**
  - Similar to SQL joins (`inner`, `outer`, `left`, `right`).
  - Requires key(s) to merge on (can be columns or indices).
  - More flexible and powerful for relational joins.

- **`join()`**
  - Convenient wrapper around `merge()` when joining on the index.
  - Less flexible than `merge()`, but simpler syntax for index-based joins.

- **`concat()`**
  - Combines DataFrames along a particular axis (`axis=0` for rows, `axis=1` for columns).
  - Does not require keys; aligns by index.
  - Useful for stacking DataFrames.

### 8. How would you handle and remove duplicate rows?

- Use `df.duplicated()` to identify duplicate rows.
- Use `df.drop_duplicates()` to remove them.
- Can specify:
  - `subset`: columns to consider for identifying duplicates.
  - `keep`: which duplicate to keep (`first`, `last`, `False` to drop all).

---

### 9. How do you efficiently filter large datasets?

- Use vectorized operations and boolean indexing for performance.
- Use `query()` for more readable, SQL-like filtering.
- Use `.loc[]` or `.iloc[]` for label- or position-based filtering.
- Avoid looping through rows (`iterrows()` is slow for large data).
- For very large datasets:
  - Use `chunking` with `pd.read_csv(..., chunksize=...)`.
  - Consider `Dask` or `Vaex` for out-of-core processing.

---

### 10. How do you change column names or reorder columns in a DataFrame?

- **Rename columns** using:
  - `df.rename(columns={'old': 'new'})`
  - `df.columns = ['new1', 'new2', ...]` to rename all columns.

- **Reorder columns** using:
  - `df = df[['col3', 'col1', 'col2']]`
  - Use list slicing or `sorted(df.columns)` for dynamic reordering.

### 11. **How do you handle time series data in Pandas?**

To handle time series data in pandas, you typically:

1. **Convert date columns to datetime type** using `pd.to_datetime()`.
2. **Set the datetime column as the index** with `df.set_index()`.
3. Use pandas time-aware functions like:
   - `resample()` for frequency conversion
   - `rolling()` for windowed operations
   - `shift()` for lagging/leading data
   - `date_range()` for generating time-based ranges
   - `dt` accessor to extract time components (year, month, day, etc.)

```python
import pandas as pd

# Convert string to datetime and set as index
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)
```

### 12. **Explain the use of `resample()`, `rolling()`, and `expanding()` functions.**

| Function       | Description                                           | Example                                |
|----------------|-------------------------------------------------------|----------------------------------------|
| `resample()`   | Resamples time series data to a new frequency         | `df.resample('M').mean()`              |
| `rolling()`    | Performs rolling window calculations                  | `df['value'].rolling(window=3).mean()` |
| `expanding()`  | Calculates cumulative values expanding over time      | `df['value'].expanding().mean()`       |

```python
# Resample to monthly frequency
monthly_avg = df.resample('M').mean()

# 3-day rolling average
rolling_avg = df['value'].rolling(window=3).mean()

# Expanding mean (cumulative average)
expanding_avg = df['value'].expanding().mean()
```

### 13. **How would you convert a string column to datetime and extract features like year, month?**

```python
df['date'] = pd.to_datetime(df['date'])  # Convert string to datetime
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['weekday'] = df['date'].dt.day_name()
```

### 15. **How do you shift data in a time series?**

```python
df['lag_1'] = df['value'].shift(1)       # Previous day's value
df['lead_1'] = df['value'].shift(-1)     # Next day's value
```

### 16. **How do you handle missing dates in a time series?**

- **Reindex** to include missing dates:
```python
full_range = pd.date_range(start=df.index.min(), end=df.index.max())
df = df.reindex(full_range)
```

- **Forward fill** or **backward fill** missing values:
```python
df.ffill(inplace=True)
df.bfill(inplace=True)
```

### 17. **What is the difference between `resample()` and `groupby()` for dates?**

- `resample()` is time-aware and works on **datetime index**.
- `groupby()` groups based on values and works on **datetime columns**.

```python
# Using resample (requires datetime index)
df.resample('M').sum()

# Using groupby (on datetime column)
df.groupby(df['date'].dt.month).sum()
```
