# 🐼 Pandas Cheat Sheet by Category

---

### Data Creation

| Function          | Description               | Example                          |
|-------------------|---------------------------|---------------------------------|
| `pd.Series()`     | Create a Series            | `pd.Series([1, 2, 3])`          |
| `pd.DataFrame()`  | Create a DataFrame         | `pd.DataFrame({'A':[1,2], 'B':[3,4]})` |

---

### Data Inspection

| Function/Attribute | Description              | Example               |
|--------------------|--------------------------|-----------------------|
| `.head()`          | First 5 rows             | `df.head()`           |
| `.tail()`          | Last 5 rows              | `df.tail()`           |
| `.info()`          | Summary info             | `df.info()`           |
| `.describe()`      | Summary statistics       | `df.describe()`       |
| `.shape`           | Shape (rows, columns)    | `df.shape`            |

---

### Selection & Indexing

| Function           | Description               | Example                   |
|--------------------|---------------------------|---------------------------|
| `.loc[]`           | Label-based indexing       | `df.loc[0, 'A']`          |
| `.iloc[]`          | Position-based indexing    | `df.iloc[0, 1]`           |
| `.at[]`            | Fast scalar access (label) | `df.at[0, 'A']`           |
| `.iat[]`           | Fast scalar access (pos)   | `df.iat[0, 1]`            |

---

### Filtering

| Method             | Description                | Example                            |
|--------------------|----------------------------|----------------------------------|
| Boolean Indexing    | Filter rows by condition    | `df[df['A'] > 5]`                 |
| `.query()`         | Query with expression       | `df.query('A > 5 and B < 3')`    |

---

### Data Manipulation

| Function           | Description                | Example                           |
|--------------------|----------------------------|---------------------------------|
| `.assign()`        | Add new columns            | `df.assign(C=df['A'] + df['B'])` |
| `.drop()`          | Drop rows or columns       | `df.drop('A', axis=1)`           |
| `.rename()`        | Rename columns or index    | `df.rename(columns={'A': 'a'})`  |
| `.sort_values()`   | Sort by column values      | `df.sort_values('B')`            |

---

### Aggregation & Grouping

| Function           | Description                | Example                             |
|--------------------|----------------------------|-----------------------------------|
| `.groupby()`       | Group data                 | `df.groupby('A').sum()`            |
| `.agg()`           | Aggregate with multiple funcs | `df.agg({'A': ['sum', 'mean']})`  |
| `.pivot_table()`   | Create pivot tables        | `df.pivot_table(index='A', values='B', aggfunc='mean')` |

---

### Missing Data Handling

| Function           | Description                | Example                           |
|--------------------|----------------------------|---------------------------------|
| `.isna()` / `.isnull()` | Detect missing values     | `df.isna()`                     |
| `.fillna()`        | Fill missing values        | `df.fillna(0)`                  |
| `.dropna()`        | Drop rows with missing data| `df.dropna()`                   |

---

### Date & Time

| Function/Attribute | Description                | Example                             |
|--------------------|----------------------------|-----------------------------------|
| `pd.to_datetime()` | Convert to datetime         | `pd.to_datetime(df['date'])`       |
| `.dt.year/month/day` | Extract datetime parts     | `df['date'].dt.year`               |
| `.resample()`      | Resample time-series data   | `df.resample('M').mean()`          |

---

### Input / Output

| Function           | Description                | Example                          |
|--------------------|----------------------------|--------------------------------|
| `pd.read_csv()`    | Read CSV file              | `pd.read_csv('file.csv')`       |
| `.to_csv()`        | Write DataFrame to CSV     | `df.to_csv('output.csv')`       |

---

### Others / Utilities

| Function           | Description                | Example                           |
|--------------------|----------------------------|---------------------------------|
| `.apply()`         | Apply function to columns/rows | `df['A'].apply(np.sqrt)`       |
| `.merge()`         | Merge/join DataFrames      | `pd.merge(df1, df2, on='key')`  |
| `.concat()`        | Concatenate DataFrames     | `pd.concat([df1, df2])`          |
| `.unique()`        | Unique values in column    | `df['A'].unique()`               |
| `.value_counts()`  | Count unique values        | `df['A'].value_counts()`         |
