### Pandas Data Structures:

Series:

* 1-dimensional labeled array.

* Homogeneous data (all elements of the same data type).

* Values are mutable, but size is immutable.

* Analogy: A single column of a spreadsheet.



DataFrame:

* 2-dimensional labeled data structure with rows and columns.

* Heterogeneous data (each column can have a different data type).

* Both size and values are mutable.

* Analogy: A spreadsheet or SQL table.


Index/MultiIndex:

* Index: Labels for rows (and columns). Can be of any data type. Allows for efficient lookups and alignment.

* MultiIndex (Hierarchical Indexing): Multiple levels of indexing for rows or columns, enabling complex data organization and slicing.

### II. Data Selection, Filtering & Manipulation (Core Skills)

Indexing and Slicing:
* loc (Label-based indexing): Select by row/column labels.
* Example: df.loc[row_label, col_label] or df.loc[start_row:end_row, ['col1', 'col2']]

* iloc (Integer-position based indexing): Select by integer positions.
* Example: df.iloc[row_index, col_index] or df.iloc[0:5, [0, 2]]

Difference between loc and iloc: Crucial for interviews. loc is inclusive of the end label for slicing, iloc is exclusive of the end position for slicing (like Python lists).

* Direct column selection: df['column_name'] or df.column_name



Conditional Filtering (Boolean Indexing): Selecting rows based on one or more conditions.

Example: df[df['Age'] > 30] or df[(df['City'] == 'New York') & (df['Salary'] > 50000)]

### Grouping and Aggregation:

groupby(): Split-Apply-Combine strategy.

Splitting data into groups based on one or more keys.

Applying a function to each group independently (e.g., mean(), sum(), count(), min(), max(), std(), size()).

Combining the results into a new data structure.

.agg(): Applying multiple aggregation functions to one or more columns within a groupby object.

.transform(): Performing a group-wise transformation that returns a result with the same shape as the input. Useful for normalization, filling missing values within groups.

V. Performance & Best Practices

Vectorization vs. Iteration:

Key takeaway: Avoid explicit loops (for loops, .iterrows(), .itertuples()) for element-wise operations as much as possible.

Emphasize using built-in Pandas methods or NumPy operations, which are highly optimized (vectorized).

Explain why vectorization is faster (underlying C/Cython implementations).

SettingWithCopyWarning:

What it is: Occurs when you try to modify a "view" of a DataFrame rather than a "copy," leading to unpredictable behavior.

How to avoid: Explicitly use .copy() when you intend to modify a subset, or use .loc for chained assignment.

Example: df[df['col'] > 0]['new_col'] = 1 (bad) vs. df.loc[df['col'] > 0, 'new_col'] = 1 (good).

Data Types Optimization:

Using appropriate data types (category, int8, float32, object for mixed types, string dtype in Pandas 2.0+).
