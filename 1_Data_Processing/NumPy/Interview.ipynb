# 📊 NumPy Interview Questions

## ✅ Core Array Operations
1. What are the differences between Python lists and NumPy arrays?
2. Explain broadcasting in NumPy with an example.
3. How does vectorization improve performance in NumPy?
4. How can you reshape a 1D array into a 2D array? What happens if the total size doesn't match?
5. What is the difference between `flatten()` and `ravel()`?

## ✅ Memory & Performance
6. How does NumPy manage memory more efficiently than native Python?
7. What is the difference between `np.copy()` and view/slicing?
8. Explain the use of `.itemsize`, `.nbytes`, and `.dtype` in memory analysis.
9. How do you optimize large matrix computations using NumPy?
10. How does NumPy internally handle array broadcasting and alignment?

## ✅ Indexing & Slicing
11. Explain advanced indexing vs boolean indexing.
12. How would you extract all even rows or columns from a 2D NumPy array?
13. What is the difference between `a[1:3]`, `a[[1, 2]]`, and `a[a > 0]`?

## ✅ Statistical & Mathematical Functions
14. How do you compute the correlation coefficient between two arrays in NumPy?
15. When would you use `np.percentile()` over `np.quantile()`?
16. How is `np.dot()` different from `np.matmul()` or `@` in NumPy?

## ✅ Random & Simulation
17. How do you generate a reproducible random array in NumPy?
18. What is the difference between `np.random.rand()`, `np.random.randn()`, and `np.random.randint()`?
19. How do you simulate a normal distribution with given mean and standard deviation?

## ✅ Array Manipulation & Transformation
20. What is the difference between `np.vstack()`, `np.hstack()`, and `np.concatenate()`?
21. How do you remove duplicate rows in a 2D NumPy array?
22. How do you efficiently remove NaNs from a NumPy array?

## ✅ Edge Cases & Best Practices
23. How would you handle operations on arrays with NaN or Inf values?
24. Why is it better to use `np.where()` or `np.select()` over Python loops?
25. Explain how broadcasting works in a real-world example, such as standardizing a dataset column-wise.
