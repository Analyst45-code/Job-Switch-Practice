Data types can be defined at creating the numpy array and converted to other types later. 

You can use either _type, type code_ or _```np``` dot_ methods to define the data type of an array, but when you use ```np``` dot method to define the data type, it can only follow _type_ rather than _type code_.

````python
arr = np.array([1,2,3], dtype='f4')
arr.dtype
````

````python
# Identical to the above
arr = np.array([1,2,3], dtype='float32')
arr.dtype
````

````python
arr = np.array([1+2j, 3-4j], dtype=np.complex64)
arr.dtype
````

````python
# Identical to the above
arr = np.array([1+2j, 3-4j], dtype='c8')
arr.dtype
````

````python
# ERROR
arr = np.array([1+2j, 3-4j], dtype='c8')
arr.dtype
````

### Type Conversion

```astype``` method: convert the data type of an array to other data types. 

Notice that ```astype``` returns a copy of the array instead of converting the data type in place. You need to assign the copy to the original array or a new array.

````python
arr = np.array([1,2,3], dtype='int16')
print('Original Data Type: ' + str(arr.dtype))

arr = arr.astype(np.float32)
print('Data Type After Conversion: ' + str(arr.dtype))
````

__WARNING__: be cautious about data overflow when you downcast the data type (from higher precision to lower precision). Some unexpected and undefined values might occur and it is usually difficult to debug such issues. 

````python
# An example of integer overflow at downcasting
arr = np.array([126,127,256], dtype='int16')
print('np array before type conversion: ' + str(arr))

# Range of int8 [-128, 127], 256 overflows after conversion
arr = arr.astype('int8')
print('np array after type conversion: ' + str(arr))
````

### String and Unicode Data Type

The ```string_``` and ```unicode_``` data types are all implicitly _fixed-length_. 

The length of the string is given by their type code appended with a number. For example, ```S3``` represents string of length 3; ```U10``` represents unicode of length 10. Otherwise, the default length is the length of the longest string in the array.

If the length of a string in the array is shorter than the length of the data type defined or converted to, the string will be truncated.

````python
# An example of truncated string
s = np.array(['abc', 'defg'], dtype='S3')
print(s)

# An example of truncated unicode
s = np.array(['abcd', 'efghi'], dtype='U3')
print(s)
````

````python
arr = np.array(['a', 'ab', 'abc'], dtype=np.str_)
print('The array is ' + str(arr))
print('The data type is ' + str(arr.dtype) + ' because the longest string in the array is "abc" and its length is 3.')

arr = np.array(['a', 'abc', 'abcd'], dtype=np.str_)
print('The array is ' + str(arr))
print('The data type is ' + str(arr.dtype) + ' because the longest unicode in the array is "abcd" and its length is 4.')
````

### 🧠 NumPy Data Types: Differences, Notes, and When to Use What

---

## 🔢 NumPy Data Types (`dtype`)

NumPy supports a variety of data types optimized for performance and memory usage.

---

### 1. **Integer Types**

| Data Type | Description | Example dtype | Range |
|-----------|-------------|----------------|--------|
| `int8`    | 8-bit signed integer | `np.int8` | -128 to 127 |
| `int16`   | 16-bit signed integer | `np.int16` | -32,768 to 32,767 |
| `int32`   | 32-bit signed integer | `np.int32` | ±2.1e9 |
| `int64`   | 64-bit signed integer | `np.int64` | ±9.2e18 |

- ✅ **Use when**: You need whole numbers.
- 📌 **Note**: Choose the smallest size that fits your data to save memory.

---

### 2. **Unsigned Integer Types**

| Data Type | Description | Example dtype | Range |
|-----------|-------------|----------------|--------|
| `uint8`   | 8-bit unsigned integer | `np.uint8` | 0 to 255 |
| `uint16`  | 16-bit unsigned integer | `np.uint16` | 0 to 65,535 |
| `uint32`  | 32-bit unsigned integer | `np.uint32` | 0 to 4.3e9 |
| `uint64`  | 64-bit unsigned integer | `np.uint64` | 0 to 1.8e19 |

- ✅ **Use when**: You don’t need negative numbers (e.g., image data).
- 📌 **Note**: More compact than signed ints if negatives are unnecessary.

---

### 3. **Floating-Point Types**

| Data Type | Description | Example dtype | Precision |
|-----------|-------------|----------------|-----------|
| `float16` | Half-precision float | `np.float16` | ~3 decimal digits |
| `float32` | Single-precision float | `np.float32` | ~7 decimal digits |
| `float64` | Double-precision float | `np.float64` | ~15 decimal digits |

- ✅ **Use when**: You need decimals (e.g., ML, physics).
- 📌 **Note**: `float64` is default. Use `float32` for performance/memory.

---

### 4. **Complex Types**

| Data Type | Description | Example dtype |
|-----------|-------------|----------------|
| `complex64`  | Complex (2 x 32-bit float) | `np.complex64` |
| `complex128` | Complex (2 x 64-bit float) | `np.complex128` |

- ✅ **Use when**: Working with signals, FFTs, quantum computing.
- 📌 **Note**: Not often used in general applications.

---

### 5. **Boolean Type**

| Data Type | Description | Example dtype |
|-----------|-------------|----------------|
| `bool_`   | Boolean value | `np.bool_` |

- ✅ **Use when**: Storing binary values (True/False).
- 📌 **Note**: Takes 1 byte per element.

---

### 6. **String Types**

| Data Type | Description | Example dtype |
|-----------|-------------|----------------|
| `str_`    | Unicode string | `np.str_` |
| `bytes_`  | Byte string | `np.bytes_` |

- ✅ **Use when**: Handling text data.
- 📌 **Note**: Not as efficient as Python strings; use `object` if needed.

---

### 7. **Object Type**

| Data Type | Description | Example dtype |
|-----------|-------------|----------------|
| `object_` | Generic Python object | `np.object_` |

- ✅ **Use when**: Mixed types or custom Python objects.
- 📌 **Note**: Slower and less efficient. Use only when required.

---

## ⚖️ Summary: When to Use What?

| Use Case              | Recommended dtype     | Notes                               |
|-----------------------|------------------------|-------------------------------------|
| Counting small values | `int8`, `uint8`        | e.g., grayscale pixel values        |
| Large integers        | `int64`                | For big datasets or unique IDs      |
| Decimal values        | `float32`, `float64`   | `float32` = performance, `float64` = precision |
| Machine learning      | `float32`              | Most frameworks use this            |
| Boolean masks         | `bool_`                | For filtering or conditions         |
| Scientific computing  | `float64`, `complex128`| Precision-critical applications     |
| Text/Labels           | `str_`, `object_`      | Use `object` for mixed types        |
| Mixed types/objects   | `object_`              | Slower but flexible                 |

---

## 💡 Notes & Tips

- **Memory Efficiency**: Choose the smallest data type that fits your data.
- **Speed**: Smaller data types often compute faster.
- **Defaults**:
  - Integers: `int64` (Linux/macOS), `int32` (Windows)
  - Floats: `float64`
- **Type Casting**: Use `.astype(dtype)` to convert between types.
- **Avoid Mixing Types**: Mixed types = slower, less optimized arrays.
