```python
import numpy as np
```

```python
arr = np.array(range(100)).reshape((10,10))
arr_1d = np.array(range(20))
```

### select an element by row and column indices

```python
print(arr[5][5])

# or more concisely
print(arr[5,5])

# Accessing a 1D array
print(arr_1d[7])

# Slicing a 1D array
print(arr_1d[5:10])  # From index 5 to 9
print(arr_1d[:5])    # From start to index 4 
print(arr_1d[5:])    # From index 5 to the end
print(arr_1d[::2])   # Every second element 

# Slicing a 2D array
print(arr[2:5, 3:6])  # Rows 2 to 4 and columns 3 to 5
print(arr[2:5, :])    # Rows 2 to 4 and all columns
print(arr[:, 3:6])    # All rows and columns 3 to 5     
print(arr[2:5, 3])    # Rows 2 to 4 and column 3
print(arr[2:5, 3:6])  # Rows 2 to 4 and columns 3 to 5      
print(arr[2:5, 3:6].shape)  # Shape of the sliced array
print(arr[2:5, 3:6].dtype)  # Data type of the sliced array

# Accessing a specific element in a 2D array
print(arr[2, 3])  # Element at row 2, column 3      

# Slicing a 2D array with a step
print(arr[::2, ::2])  # Every second row and every second column    

# Accessing a specific row or column
print(arr[5, :])    # Row 5   
print(arr[:, 5])    # Column 5

# Accessing a specific row or column with a step
print(arr[::2, 5])  # Every second row in column 5  

# Accessing a specific row or column with a step
print(arr[5, ::2])  # Row 5 and every second column
```

### indexing with slicing

```python
print(arr[1:3, 4:6])
```

```python
# ellipsis slicing: auto-complete the dimensions
arr = np.array(range(16)).reshape(2,2,2,2)

# equivalent to arr[0,:,:,:]
print(arr[0, ...])    
```

### assign a scalar to a slice by broadcasting

```python
arr[1:3,:] = 100    # or simply arr[1:3]
arr[:,8:] = 100
print(arr)
```

### boolean indexing

```python
arr1 = np.arange(25).reshape((5,5))

bools = np.array([True, True, False, True, False])
print(arr1[bools])
```

```python
# negate the condition
print(arr1[~bools])    
```

```python
arr2 = np.array([1,2,3,4,5])
# multiple conditions
print(arr1[(arr2<2) | (arr2>4)])    
```

### fancy indexing

```python
arr = np.random.rand(10,10)
```

```python
# select arr[3,3], arr[1,2], arr[2,1]
print(arr[[3,1,2], [3,2,1]])       
```

```python
# select rows 3,1,2 and columns 6,4,8 
print(arr[[3,1,2]][:, [6,4,8]])    
```

### dimension inference

```python
# dimension inference using any negative number (usually -1)
arr = np.array(range(16)).reshape((4,-1))
print(arr.shape)
```

### find elements/indices by conditions

```python
arr = np.arange(16).reshape(4,4)
```

```python
# find the elements greater than 5 and return a flattened array
print(arr[arr>5])    # or arr[np.where(arr>5)]
```

```python
# return values based on conditions 
# np.where(condition, true_return, false_return)
print(np.where(arr>5, -1, 10))
```

```python
# find the indices of the elements on conditions
print(np.argwhere(arr>5))
```
